### World Heatmap of American Gut Samples 

The mapping file is located at the following address:

ftp://ftp.microbio.me/AmericanGut/manuscript-package/1250/ag_map_with_alpha.txt

It can be downloaded by right clicking and selecting "Save As" after navigating to the page


For setting up the conda enviornment, the packages <b>folium</b> and <b>geopandas</b> are needed. A .yml file titles "Folium_Heatmaps.yml" is also provided with the full enviornment. 




```python
import pandas as pd 
import folium
from folium.plugins import HeatMap
```

```python
mapping = pd.read_table('ag_map_with_alpha.txt', index_col=0, low_memory=False)
```

```python
#Drop cases where latitude and longitude are missing
mapping = mapping[mapping['latitude']!='Missing: Not collected']
mapping = mapping[mapping['latitude']!='Unspecified']

mapping = mapping[mapping['longitude']!='Missing: Not collected']
mapping = mapping[mapping['longitude']!='Unspecified']
```

```python
#Make latitude and longitude into float
mapping['longitude'] = mapping['longitude'].astype(float)
mapping['latitude'] = mapping['latitude'].astype(float)
```

```python
lat_long = mapping[['latitude', 'longitude']]
```

```python
#Make the latitude and longitude into tuples for folium
tuples = [tuple(x) for x in lat_long.values]
```

```python
#Create the heatmap. The map is interactive; hold left click to drag and scroll bar to zoom

heatmap = folium.Map([20., 10.], tiles='Mapbox Bright', zoom_start=2)

heatmap.add_children(HeatMap(tuples, radius=8, blur=8, min_opacity=0.5))

heatmap
```
